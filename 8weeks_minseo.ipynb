 데이터 분석 이란
-----
-알고리즘,수학적 지식을 적용해 데이터에 대한 인사이트를 도출하는 것

데이터 분석과정
----
### 1.탐색적 데이터 분석 (Exploratory Data Anlysls)

     -현재 문제상황 정의
     -시각화,변수탐색
     -결측치와 이상치 탐지
     
### 2. 데이터 전처리(Data preprocessing)

    -문제해결을 위한 적적한 데이터 처리
    -정형화된 형태로 데이터를 변화시킴
    
### 3. Feathre Engine
    
    -변수의 추가,축소
    -특징 추출
    
### 4. 모델링(modeling)

    -정형화된 데이터로
    -데이터를 통해 예측,분류,결과해석
    
### 데이터 자체가 분석에서 가장 중요  

    -Data preprocessing과정에 대부분 시간 소요

 ### 2. Matplotlib, Seaborn
 
     Matplotlib
   - 기본적인 시각화 도구
   
   >>그래프 종류
    1. 산점도 scatter chart
        2개의 연속형(숫자로 연속적으로 표현한)데이터 표현에 적합
    
    2. 막대그래프 bar chart
        하나의 데이터가 가지는 빈도표현에 적합

    3. 히스토그램  histogram
        데이터가 특정 구간사이에 가지는 빈도를 세밀하게 표시 가능 
    
    4. 원그래프 pir chart
        데이터가 전체에서 가지는 비율을 한눈에 보는데 적합
    
    5. 박스그래프 box plot
        데이터속 이상치나 기술통계량 값을 보는데 적합
        
    6. 껵은선 그래프 Line plot
        단위흐름에 자료의 양을 선으로 그린 그래프
    .
    .
    .
     
    Matplotlib 기본적인 시각화 + Seeborn 색상,통계용 차트
    
    #두 변수의 관계를 표현하는 jointplot
    
    #산점도 그리기
    jointplot(x,y,kind=scatter,legand.....)
    
    #산점도에 회귀선을 넣어 그래프 그리기
    jointplot(x,y,kind='reg',legand....)
    
    >>회귀선이란? 
    주어진 데이터로 직선을 그려 데이터와 직선의 차이값의 제곱이 최소가 되는 직선
    
    # resid 그래프 두 변수의 잔차를 보여줌
    jointplot(kind="resid",...)
    
    # hex 그래프 
    육각형으로 데이터가 많이 몰려있는 부분을 더 진하게 표시
    별로 안쓴다
    
    #데이터 프레임속 표현가능한 모든 변수들끼리 1대1로 관계를 표현해주는 pairplot
  
###상관성 분석(Correlation Analysis)
    양적 변수들간의 상관관계를 -1 ~ 1까지의 수로 표현
    1에 가까울수록 두 변수끼리 강한 선형관계를 지님
    
### 이상치
    시각화를 통해 이상치를 표현(box plot)
    
    자연발생한 이상치가 아닌것 이라 판단될 경우
        1. 삭제
        2. 대체
        >> 평균,중앙,최빈값중 하나로 대체
        >> 예측모델을 만들어 예측된 값으로 대체
      
    자연발생한 이상치라 판단될 경우
        1.케이스 분리해석 
        2.변수를 리샘플링
        3.이상치를 설명 가능한 새로운 변수생성
