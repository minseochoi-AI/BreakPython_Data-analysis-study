```python
# 남북한발전전력량.xlsx 파일엔 남한과 북한의 발전전력별 연간통계량이 저장되어있다.
# 해당 데이터를 읽어와서 남한과 북한의 수력 및 화력발전과 그 합계에 대한 연간 통계량만을 분리해서 남북한수력및화력발전전력량.csv로 저장하라.

# 남북한 발전전력 데이터 읽어옴
eng_df=pd.read_excel(r"C:\남북한발전전력량.xlsx")

# 데이터 프레임의 행,열을 교체(전치행렬화)
eng_df=eng_df.transpose()

# 남북,에너지 관련 컬럼 모두 삭제
eng_df=eng_df.drop([eng_df.index[0],eng_df.index[1]])

# 국가,발전종류가 모두 담긴 컬럼이름 리스트 생성
eng_columns=["남한 합계","남한 수력","남한 화력","남한 원자력","남한 신재생","북한 합계","북한 수력","북한 화력","북한 원자력",]

# 새롭게 만든 컬럼명을 대입
eng_df.columns=eng_columns


# 데이터 프레임 속 값이 없는 부분을 0으로 대체
eng_df=eng_df.replace("-",0)

# 남한 데이터 중 수력,화력 발전량의 합계만 구하기 위해 원자력,신재생 등의 컬럼값을 합계에서 뻄
# 남한 합계 발전량 구하기
n_eng_sum=eng_df['남한 합계']-(eng_df['남한 원자력']+eng_df['남한 신재생'])

# 바뀐 합계 대입
eng_df['남한 합계']=n_eng_sum

# 필요없는 컬럼 삭제

eng_df=eng_df.drop(['남한 원자력','남한 신재생','북한 원자력'],axis=1)

eng_df.to_csv("final_eng_df.csv",encoding="utf-8-sig")
```

```python
# 다음을 만족하는 20년 달력 데이터 프레임을 만드시오
# original    Year    Month    Day    weekday    weekend    season    
# 0   20-01-01     20       1       1       Wed         0       winter
# 1   20-01-02     20       1       2       Thr         0       winter
# 2   20-01-03     20       1       3       Fri         0       winter
# 3   20-01-03     20       1       4       Sat         1       winter

# ...   ...        ...     ...     ...      ...        ...        ...   

# 365 20-12-31     20       12      31      Thr         0       winter

# 컬럼 이름을 변수로 저장
year_columns=["original","Year","Month","Day","weekday","weekend","season"]   

# 시간 모듈 임포트
import datetime

# for 문으로 전체날짜 정보(original) 생성을 위해 2020,1,1 날짜 생성
original_data=datetime.date(2020,1,1)

# original 컬럼에 들어갈 정보를 넣을 리스트 생성
original_list=[]

#  year 컬럼에 들어갈 정보를 넣을 리스트 생성
year_list=[]

# month 컬럼에 들어갈 정보를 넣을 리스트 생성
month_list=[]

# original_data에 저장 된 날짜 정보중 월정보를 가져올 month_data 생성(월,계절 정보 구하는데 사용)
month_data=""

# day 컬럼에 들어갈 정보를 넣을 리스트 생성
day_list=[]

# weekday컬럼에 들어갈 정보를 넣을 리스트 생성
weekday_list=[]

# 요일정보를 weekday에 저장
# 주어진 예시에 따라 2020-1-1이 wed 요일이 될 수 있도록 요일저장
weekday=["Mon","Tue","Wed","Thr","Fri","Sat","Sun"]

# 요일정보를 weekday함수를 통해 숫자로 받아올 week_data생성(요일,주말 정보 구하눈데 사용)
week_data=""

# weekend컬럼에 들어갈 정보를 넣을 리스트 생성
weekend_list=[]

# season 컬럼에 들어갈 정보를 넣을 리스트 생성
season_list=[]

# 365일 날자 정보를 얻기 위해 위해 365번 반복
for i in range(0,366):

    # original_data정보를 origianl_list에 저장
    original_list.append(original_data)

    # year 컬럼에 20숫자를 계속 추가
    year_list.append(20)

    # 1.orignal_data정보에 있는 달 정보를 strftime 함수를 사용해 month_data에 추가
    # month_list에 month_data 추가
    month_data=original_data.strftime('%m')
    month_list.append(month_data)

    # day리스트에 orignal_data정보에 있는 일 정보를 strftime 함수를 사용해 day_list에 추가
    day_list.append(original_data.strftime('%d'))

    # 1. orignal_data정보에 있는 날짜 정보를 weeday함수로 요일을 나타내는 숫자형태로 변환
    # 2. 이를 week_data에 저장 
    # 3. 저장한 숫자를 weekday에 저장해논 요일을 인덱싱 하는데 사용 
    # 3. 인덱싱으로 찾은 정보를 weekday_list에 추가
    week_data=original_data.weekday()
    weekday_list.append(weekday[week_data])
    
    # if문을 사용해 week_data 변수에서 주말을 나타내는 6숫자가 나올경우 weekend_list에 1을 그럴지 않을경우 0을 추가
    if week_data==6:
        weekend_list.append(1)
    else:
        weekend_list.append(0)

    # month_data에 저장된 정보를 if문으로 비교하기 위해 타입을 str에서 int형으로 바꿈
    month_data=int(month_data)

    # 만약 month_data가 3~5에 해당한다면
    if (month_data >= 3) & (month_data <= 5):
        # season_list에 "Spring"추가
        season_list.append("Spring")

    # 만약 month_data가 6~8에 해당한다면    
    elif (month_data >= 6) & (month_data <= 8):
        # season_list에 "Summer"추가
        season_list.append("Summer")

    # 만약 month_data가 9~10에 해당한다면  
    elif (month_data >= 9) & (month_data <= 10):
        # season_list에 "Summer"추가
        season_list.append("Fall")

    # month_data가 이외의 숫자(1,2,11,12)에 해당한다면  
    else:
        # season_list에 "Winter"추가
        season_list.append("Winter")

    # 반복 1회마다 original_data 날짜 정보가 1일식 늘어남
    original_data=original_data+datetime.timedelta(days=1)

year_df=pd.DataFrame({"original":original_list,"Year":year_list,"Month":month_list,"Day":day_list,"weekday":weekday_list,"weekend":weekend_list,"season":season_list}) 
    
print(year_df)


# 데이터 프레임 정의
year_df=pd.DataFrame()

```
