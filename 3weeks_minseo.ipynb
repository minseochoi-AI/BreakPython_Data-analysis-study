```python
# randem 모듈 임포트
from random import*

# 함수선언
def lotto_maker(): 
# 로또숫자가 들어갈 리스트 생성
    lotto=[]
# 로또 번호 개수만큼 6번 반복
    for i in range(0,6):
# lotto 리스트에 1~45중 랜덤한 수를 대입
        lotto.append(randint(1,45))
# 만들어진 로또 번호를 정렬
    lotto.sort()
# 출력
    print(lotto)
# 함수호출
lotto_maker()
```

```python
# 함수선언
def str_analysis():
# 사용자로부터 문자열 입력
    try:
        str1=input("문자열 입력")
# 문자개수 출력
        print(f"문자의 개수:{len(str1)}")
# 가장 큰 문자열 출력
        print(f"가장 큰 문자열:{max(str1)}")
# 뒤집어 출력
        print(f"뒤집은 문자열:{str1[::-1]}")
# 문자열에 아무것도 입력하지 않은 경우
    except ValueError:
        print("값을 입력하지 않았습니다")
        exit()

# 함수호출
str_analysis()
```

```python
# datetimr 모듈 임포트
import datetime
# 함수선언
def time_read():
# 현재시간 생성
    now=datetime.datetime.now()
# 현재시간 출력
    print(f"{now.year}년{now.month}월{now.day}일{now.hour}시{now.minute}분{now.second}초")
# future변수에 49일 1시간 8분 7초후의 시간을 저장
# 49일 +

    future=now+datetime.timedelta(days=49)
# 1시간 +
    future=future+datetime.timedelta(hours=1)
# 8분 +
    future=future+datetime.timedelta(minutes=8)
# 7초+
    future=future+datetime.timedelta(seconds=7)
# future 출력
    print(f"{future.year}년{future.month}월{future.day}일{future.hour}시{future.minute}분{future.second}초")
    

    
# 함수호출
time_read()
```

```python
# 정규표현식 모듈 임포트
import re
# 함수선언
def Check_mali():
# 메일 input
    mali=input("메일을 입력하세요")
# check에 정규표현식으로 조건을 담은 객체 저장
# 계정
# 1.모든 영문,숫자 +-_.외 나머지를 제외한다=[^~!@#$%^&*()><?]
# 2.+ 1회이상 반복되어
# 3.@문자 앞까지 이어진다
# 도메인
# 4.@ 이후 [\w]모든 영문,숫자 | 또는 +-_. 특수기호가
# 5. + 1번 이상 반복된다
# 최상위 도메인
# 영어소문자,대문자,숫자만 포함해 여러단계.기호가 여러개 가능

    check=re.compile('[^~!@#$%^&*()><?]+@[^~!@#$%^&*()><?]+[\w.]*')
                    # [     계정      ] @[   도메인       ][최상위 도메인]

    c_mail=check.search(mali)
    if c_mail:
        print("올바른 형식입니다")

    else:
        print("올바르지 않습니다")
    
Check_mali()
```

```python

# 사전에 들어갈 정보를 저정할 변수,딕셔너리 생성
key=0
value=0
dict1={}
# 사용자가 종료 메뉴를 입력 할 떄가지 반복
while True:
# inp_menu변수에 사용자로부터 어떤 메뉴를 사용할지 정보를 저장함
    inp_menu=input("메뉴를 선택해주세요\n1.입력\n2.전체출력\n3.영어로검색\n4.한글로검색\n5.종료\n")
# 사용자가 입력메뉴를 선택 할 경우
    if inp_menu=="입력":
        while True:
# 예외처리 구문
            try:
# inp 변수에 영단어:한글단어 쌍으로 입력된 정보 저장
                inp=input("영단어:한글단어 쌍으로 입력해주시고, 끝났다면 나가기를 입력해주세요.")
# 만약 inp변수가 나가기 일경우
                if inp=="나가기":
                    print("사전정보 입력을 종료합니다")
# while문 탈출          
                    break
# 사전에 들어갈 정보를 입력할 경우
                else:
# inp에 들어있는 값을 key,value에 :를 기준으로 split하여 저장하고
                    key,value=inp.split(":")
# dict1에 key:value 형태로 저장한다
                    dict1[key]=value
# 입력된 정보에 :문자가 없어 split가 불가능 할때
            except ValueError:
# 오류출력 및 반복 건너뜀
                print("반드시 영문:한글 형태로 입력하셔야 합니다. 사전정보를 다시 입력하세요")
                continue
# 정의되지 않은 오류 발생시
            except Exception:
# 오류출력 및 반복 건너뜀
                print("정의되지 않은 오류가 발생했습니다.사전정보를 다시 입력하세요")
                continue
# 사용자가 종료메뉴를 선택 할 경우
    if inp_menu=="종료":
        print("사전을 종료합니다")
        exit()

# 사전에 아무런 정보가 입력되지 않은 상태에서
    if dict1=={}:
# inp_menu에 전체출력,영어검색,한글검색 명령이 입력 경우
        if (inp_menu=="전체출력") |(inp_menu=="한글로검색") |(inp_menu=="영어로검색"):
# 오류출력
            print("사전에 아무런 정보가 입력되지 않았습니다 입력을 먼저 수행해 주세요.")
        else:
            print("메뉴 이외에 다른 명령을 입력했습니다")
# 사전에 정보가 입력되어 있을 경우
    if dict1!={}:
        if inp_menu=="전체출력":
# 전체출력 메세지 출력
            print("전체출력을 선택하셨습니다")
# 사전정보 출력   
            print(dict1)
# 사용자가 영어로 검색 메뉴를 입력할 경우
        elif inp_menu=="영어로검색":
# 반복실핼
            while True:
# 예외처리 구문
                try:
# 사용자로부터 영단어를 입력받은 뒤
                    e_input=input("영단어를 입력하세요")
# 만약 나가기를 입력 할 경우
                    if e_input=="나가기":
# 반복종료
                        print("영단어 입력을 종료합니다")
                        break
# 영단어를 입력할 경우
                    else:
# 영단에에 해당하는 한글 출력
                        print(dict1[e_input])
# 입력한 key에 해당하는  value를 찾지 못한 경우
                except KeyError:
# 오류출력
                    print("입력하신 영문에 해당하는 한글을 찾지 못했습니다")
# 아무것도 입력하지 않은경우
                except ValueError:
# 오류출력
                    print("영문이 입력되지 않았습니다")
# 그외에 오류가 발생한 경우
                except Exception:
#오류출력
                    print("예상치못한 오류가 발생했습니다") 
# 사용자가 한글로검색을 입력한 경우
        elif inp_menu=="한글로검색":
# 예외처리용 구문
            while True:
# 반복실핼
                try:
# 사용자로부터 한글을 입력받은 뒤
                    k_input=input("한글을 입력하세요")
# 나가기를 입력 할 경우
                    if k_input=="나가기":
# 반복종료
                        print("한글 입력을 종료합니다")
                        break
# 한글을 압력 할 경우
                    else:
# 딕셔너리의 key,value값 교체
                        dict1={key:value for value,key in dict1.items()}
# 한글 해당하는 영문 출력
                        print(dict1[k_input])
# 츨력한 뒤 다시 원래형태로 전환
                        dict1={key:value for value,key in dict1.items()}
# 입력한 key에 해당하는  value를 찾지 못한 경우
                except KeyError:
# 오류출력
                    print("입력하신 한글에 해당하는 영문을 찾지 못했습니다")
# 아무것도 입력하지 않은경우
                except ValueError:
# 오류출력
                    print("한글이 입력되지 않았습니다")
# 그외에 오류가 발생한 경우
                except Exception:
#오류출력
                    print("예상치못한 오류가 발생했습니다")
# 사용자가 메뉴 이외 다른 명령을 입력한 경우 
    elif inp_menu!={i for i in ["입력","전체출력","영어로검색","한글로검색","종료"]}:
# 오류출력
        print("헤당되는 메뉴가 없습니다 다시 메뉴를 입력해주세요")
```
