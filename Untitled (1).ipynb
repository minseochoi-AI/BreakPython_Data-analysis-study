```python
#결측치 확인 및 제거
mark_df["Income"].fillna(mark_df["Income"].mean(), inplace=True)
mark_df.isnull().sum()
```

```python
#다항회귀 분석 income으로 나머지 구매량 예측하기
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures 

#독립변수 x를 income 종속변수 y를 나머지 구매량으로 둠
X=mark_df[["Income"]]# 2차원 array
X_train, X_test= train_test_split(X, test_size=0.3, random_state=7) #X train,teat data 분류

#list1에 들어간 컬럼 수만큼 반복
for i in range(len(list1)):
    Y=mark_df[list1[i]] #반복시마다 Y에 순차적으로 다른 컬럼값이 들어감
    Y_train, Y_test= train_test_split(Y, test_size=0.3, random_state=7) #y train,teat data 분류 난수값 행운의 7
    
    # 단순회귀분석 모형 객체 생성
    slr = LinearRegression()

    # train data를 가지고 회귀모델 학습
    slr.fit(X_train, Y_train)

    # 학습을 마친 모형에 test data를 적용하여 결정계수(R^2) 계산
    slr_r_square = slr.score(X_test, Y_test)

    Y_hat = slr.predict(X)
    
    # 회귀식과 결정계수(R^2) 산출
    print('단순회귀모델 회귀식 :', float(slr.coef_),'X +',slr.intercept_)
    print('단순회귀모델 결정계수(R^2) :', slr_r_square)
    
    plt.figure(figsize = (10, 5))
    ax1 = sns.distplot(Y, hist = False, label = "Y")
    ax2 = sns.distplot(Y_hat, hist = False, label = "Y_hat", ax = ax1)
    plt.legend()
    plt.show()
    plt.close()
    
    
    
    # 다항식 변환 
    poly = PolynomialFeatures(degree = 2) #2차항 적용
    X_train_poly = poly.fit_transform(X_train) #X_train 데이터를 2차항으로 변형

    # train data를 가지고 모형 학습
    plr = LinearRegression()   
    plr.fit(X_train_poly, Y_train)
    
     # 학습을 마친 모형에 test data를 적용하여 결정계수(R^2) 계산
    X_test_poly = poly.fit_transform(X_test) # X_test 데이터를 2차항으로 변형
    plr_r_square = plr.score(X_test_poly,Y_test)
    print('다항회귀모델 회귀식 :', plr.coef_[1],'X^2 +',plr.coef_[2],'X +',plr.intercept_)
    print('다항회귀모델 결정계수(R^2) :', plr_r_square)

    plt.figure(figsize = (10, 5))
    ax1 = sns.distplot(Y, hist = False, label = "Y")
    ax2 = sns.distplot(Y_hat, hist = False, label = "Y_hat", ax = ax1)
    plt.show()
    plt.close()
    
    # k=int(input("연간소득[Income]을 입력 해주세요"))
    #print("다항회귀모델러 예측한 당신의 와인 구매량은 \n"+((plr.coef_[2]*(k**))+(plr.coef_[1]*k)+,plr.intercept_))+"입니다")
```

```python
#다중회귀 분석

#모든 구매량 컬럼값으로 사용자의 income을 예측함
X = mark_df[["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds"]] #독립 변수 X, 독립변수는 해당 라이브러리에서 2D Array로 집어넣어야 함
Y = mark_df['Income']  #종속 변수 Y

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=7)


# 단순회귀분석 모형 객체 생성
mlr = LinearRegression()

# train data를 가지고 회귀모델 학습
mlr.fit(X_train, Y_train)

# 학습을 마친 모형에 test data를 적용하여 결정계수(R^2) 계산
r_square = mlr.score(X_test, Y_test)

# 회귀식과 결정계수(R^2) 산출
print('회귀식 :', mlr.coef_,'Xn +',mlr.intercept_)
print('결정계수(R^2) :', r_square)

Y_hat = mlr.predict(X)

plt.figure(figsize = (10, 5))
ax1 = sns.distplot(Y, hist = False, label = "Y")
ax2 = sns.distplot(Y_hat, hist = False, label = "Y_hat", ax = ax1)
plt.legend()
plt.savefig('Customed1 Plot.png')
```
